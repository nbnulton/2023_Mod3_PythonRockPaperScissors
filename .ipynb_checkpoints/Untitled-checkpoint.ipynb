```python
# computer, player, rock, paper, scissors, win, lose, randomint

from random import randrange

# move = ["rock", "paper", "scissors"]

# # get random index number
# random_move = randrange(len(move))
# # Select item using index number
# print("Computer played:", move[index])

class Game():
#                                         list
    def __init__(self, move):
        self.move = move
        
    def runGame(self, move):
        player_prompt = input("What move would you like to play? rock, paper, or scissors? ")
        random_move = randrange(len(self.move))
        if player_prompt == "rock":
            if random_move == 0:
                print("Computer played: rock")
                print("It's a tie.")
            elif random_move == 1:
                print("Computer played: paper")
                print("You lose!")
            elif random_move == 2:
                print("Computer played: scissors")
                print("You win!")
        elif player_prompt == "paper":
            if random_move == 1:
                print("Computer played: paper")
                print("It's a tie.")
            elif random_move == 2:
                print("Computer played: scissors")
                print("You lose!")
            elif random_move == 0:
                print("Computer played: rock")
                print("You win!")
        elif player_prompt == "scissors":
            if random_move == 2:
                print("Computer played: scissors")
                print("It's a tie.")
            elif random_move == 0:
                print("Computer played: paper")
                print("You lose!")
            elif random_move == 1:
                print("Computer played: paper")
                print("You win!")

        

plays = Game(["rock", "paper", "scissors"])

def run():
    while True:
        prompt = input("Would you like to play rock, paper, scissors? yes or no ")
        if prompt == "yes":
            plays.runGame("")
        elif prompt == "no":
            break

run()
```
